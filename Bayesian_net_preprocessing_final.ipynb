```python
DAG = bn.structure_learning.fit(df_sa_geo_fin_bn, methodtype='ex', scoretype='bic')
# Plot detected DAG
G = bn.plot(DAG)

# Compute edge strength using chi-square independence test
model1 = bn.independence_test(DAG, df_sa_geo_fin_bn, alpha=0.05, prune=False)
bn.plot(model1, pos=G['pos'])
```

```python
# Compute edge strength using chi-square independence test and remove (prune) the not-signficant edges
model3 = bn.independence_test(model1, df_sa_geo_fin_bn, alpha=0.05, prune=True)
bn.plot(model3, pos=G['pos'])
```

```python
# Parameter learning
model = bn.parameter_learning.fit(model3, df_sa_geo_fin_bn)
```

```python
# Print CPDs
bn.print_CPD(model)
```

```python
# professor's technique
from pgmpy.estimators import HillClimbSearch
from pgmpy.estimators import K2Score, BicScore, BDeuScore


hc = HillClimbSearch(df_sa_geo_fin_bn)


best_model = hc.estimate(scoring_method=BicScore(df_sa_geo_fin_bn))
print(best_model.edges())
```

```python
import networkx as nx
import pylab as plt
nx.draw(best_model, with_labels=True)
plt.savefig('model.png')
plt.show()
plt.close()
```
